### Análisis Descriptivo Accidentes

En el presente código se realiza un análisis descriptivo de los más de 4 millones de accidentes que se tienen registrados en la base de datos a trabajar.

```python
#======================================#
#     0. Entrada de los datos          #
#======================================#
import pandas as pd

## Directorio
dir = "C:\\Users\\ghama\\Desktop\\Minería de Datos\\"

## Entrada de los datos
dd = pd.read_csv(dir + "3 - Entregas\\0 - Insumos\\US_Accidents_depurado_v1.csv", sep = ";")

dd.head()
```

### Matriz de correlación

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.pylab as plt

variables = ["Temperature(F)","Wind_Chill(F)","Humidity(%)","Precipitation(in)","Distance(mi)","Pressure(in)"
             ,"Visibility(mi)","Wind_Speed(mph)"]

plt.close()
sns.set(rc={'figure.figsize':(12, 9)})
matriz = sns.heatmap(dd[variables].corr().round(decimals = 2)
            , annot = True
            , vmin = -1
            , vmax = 1
            , cmap = "RdBu")

plt.setp(matriz.get_xticklabels(), rotation=45, ha="right", rotation_mode = "anchor")
matriz.tick_params(direction='out', pad = 90, axis = "y")
matriz.set_yticklabels(ha = "left", labels = variables)

# matriz.set_xticklabels(ha = "right", labels = variables)
plt.title("Matriz de Correlación\n")
#plt.savefig("Otros\\matriz_corr_num.png")
#plt.close()
```

### Boxplots bivariados

```python
### Obtener variables en escala logaritmica
import numpy as np

var_y = ["Accident_duration","Interest_Points","Road_Features","Traffic_Signals","Wind_Speed(mph)","Precipitation(in)"
         ,"Wind_Chill(F)","Humidity(%)","Pressure(in)","Visibility(mi)","Distance(mi)","Temperature(F)"]

for i in var_y:
    dd["log_" + i] = np.log(dd[i])

dd.head()
```

```python
## Boxplots generales para encontrar las relaciones pertinentes
var_x = ["Severity","Side","State","Sunrise_Sunset", "Wind_Direction", "Weather_Condition",
         "Year","Month","Day","Weather_Condition_Cat"]

var_y = ["Accident_duration","Interest_Points","Road_Features","Traffic_Signals","Wind_Speed(mph)","Precipitation(in)"
         ,"Wind_Chill(F)","Humidity(%)","Pressure(in)","Visibility(mi)","Distance(mi)","Temperature(F)"]

var_y_log = ["log_" + i for i in var_y]

for i in var_x:
    for j in var_y_log:
        sns.boxplot(data = dd, x = i, y = j,  palette = "RdBu") # hue = "smoker",
        plt.savefig("Boxplots\\box_" + i + "-" + j + ".png")
        plt.close() 
```
